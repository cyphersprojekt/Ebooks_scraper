```python
#Importar los módulos necesarios para el funcionamiento de nuestro programa

import requests
import pandas as pd
```

```python
#cURL(cmd) Request del website convertida a Python utilizando curl.trillworks.com

headers = {
    'sec-ch-ua': '^\\^Google',
    'Referer': 'https://www.ebooks.com/en-ar/subjects/computers/',
    'DNT': '1',
    'sec-ch-ua-mobile': '?0',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36',
    'sec-ch-ua-platform': '^\\^Windows^\\^',
    'Content-Type': 'application/json',
}
params = (
    ('subjectId', '13'),
    ('pageNumber', '1'),
    ('countryCode', 'AR'),
)
response = requests.get('https://www.ebooks.com/api/search/subject/', headers=headers, params=params)
```

```python
#Visualizar el JSON de la variable "response" creada por el step anterior

response.json()
```

```python
#Visualizar los datos almacenados en la clave 'books' del diccionario response

response.json()['books']
```

```python
#Definir una variable para acceder más fácil a la información del step anterior

resultados_json = response.json()['books']
```

```python
#Crear listas vacías con nombre de los datos que queremos obtener de la API

titulo = []
subtitulo = []
autor = []
editorial = []
ano = []
precio = []
```

```python
#Añadir a cada una de esas listas los resultados en su respectivo orden mediante un loop

for resultados in resultados_json:
    titulo.append(resultados['title'])
    subtitulo.append(resultados['subtitle'])
    autor.append(resultados['authors'][0]['author_name'])
    editorial.append(resultados['publisher'])
    ano.append(resultados['publication_year'])
    precio.append(resultados['price'])
```

```python
#Crear un DataFrame para visualizar si nuestro output es correcto

ebooks_dataframe = pd.DataFrame({'Titulo':titulo,
                         'Subtitulo':subtitulo,
                         'Autor':autor,
                         'Editorial':editorial,
                         'Año':ano,
                         'Precio':precio})
```

```python
#Acceder al DataFrame que creamos en el step anterior

ebooks_dataframe
```

```python
#Exportar nuestros resultados a un archivo Excel

ebooks_dataframe.to_excel('EbooksDotCom.xlsx', index=False)
```
